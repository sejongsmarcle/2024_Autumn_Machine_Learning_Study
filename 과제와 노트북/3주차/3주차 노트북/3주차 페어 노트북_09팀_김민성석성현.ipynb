**연습문제 1번**

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784',as_frame=False)
X,y=mnist.data,mnist.target
X
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = X[:60000],X[60000:],y[:60000],y[60000:]
```

```python
# KNeighborsClassifier 분류기
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier() #모델 생성
knn.fit(X_train,y_train) #모델 훈련
train_accuracy = knn.score(X_train,y_train) #훈련세트 정확도 확인
print(train_accuracy)
```

```python
#그리드 서치를 통해 n_neighbors와 wights값을 최적화 해보기
from sklearn.model_selection import GridSearchCV
import numpy as np
```

```python
#파라미터 그리드 설정
param_grid={
    'n_neighbors':np.arange(3,10), # 3~9까지 이웃수 탐색
    'weights':['uniform','distance'] # 가중치는 균등or거리 기반
}
```

```python
# 그리드 탐색 설정
grid_search=GridSearchCV(KNeighborsClassifier(),param_grid,cv=2,n_jobs=-1,verbose=2)
```

```python
# 그리드 탐색 실행
grid_search.fit(X_train,y_train)
```

```python
#하이퍼파라미터 출력
grid_search.best_params_

#최적의 파라미터로 모델 생성
best_knn=grid_search.best_estimator_

#최적화 된 모델로 정확도 계산
test_accuracy=best_knn.score(X_test,y_test)
test_accuracy
```
