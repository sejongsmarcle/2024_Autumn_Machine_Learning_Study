## 1. MNIST 데이터셋으로 분류기를 만들어 테스트 세트에서 97% 정확도를 달성해보세요.

```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', as_frame=False)
```

```python
X, y = mnist.data, mnist.target
X
```

```python
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
X_train
```

```python
y_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier() #KNN import
```

```python
knn_clf.fit(X_train, y_train) #Knn 모델 학습함
```

```python
baseline = knn_clf.score(X_test,y_test) # 성능측정
```

```python
baseline
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier() # RandomForest import(다른 모델 테스트)
rfc.fit(X_train, y_train)
```

```python
baseline_rfc = rfc.score(X_test,y_test)
```

```python
baseline_rfc # 비슷한 정확도이므로 사용 가능
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = [{'weights' : ["uniform","distance"], 'n_neighbors' : [3, 4, 5, 6]}]
```

```python
knn_clf = KNeighborsClassifier()
```

```python
grid_search = GridSearchCV(knn_clf, param_grid, cv = 5)
grid_search.fit(X_train[:10000], y_train[:10000])
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
grid_search.best_estimator_.fit(X_train, y_train)
tuned_accuracy = grid_search.score(X_test, y_test)
tuned_accuracy
```

```python
knn = KNeighborsClassifier(n_neighbors=4, weights='distance')
knn.fit(X_train, y_train)
```

```python
accuracy = knn.score(X_test,y_test)
accuracy
```

```python
param_grid = [{"criterion" : ['gini','entropy','log_loss']}]
grid_search = GridSearchCV(rfc, param_grid, cv = 5)
grid_search.fit(X_train[:10000], y_train[:10000])
```

```python
rfc2 = RandomForestClassifier(criterion='entropy')
rfc2.fit(X_train,y_train)
acr = rfc2.score(X_test,y_test)
acr
```
