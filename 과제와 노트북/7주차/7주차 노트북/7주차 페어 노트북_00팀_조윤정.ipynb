```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import fetch_openml

#데이터 로드
mnist = fetch_openml('mnist_784', version=1)
X, y = mnist.data, mnist.target

print(X.shape)
print(y.shape)
```

```python
X_train, X_vaild, X_test, y_train, y_vaild, y_test = X[:50000], X[50000:60000], X[60000:], y[:50000], y[50000:60000], y[60000:]
```

# 3주차 내용

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score


# K-최근접 이웃 분류기 생성
knn = KNeighborsClassifier(n_neighbors=3, weights = 'distance', metric = 'manhattan')

# 학습
knn.fit(X_train, y_train)

# 예측
y_pred = knn.predict(X_test)

# 정확도 평가
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# KNN 분류 모델을 사용한다고 가정
model = KNeighborsClassifier()

# 탐색할 하이퍼파라미터의 그리드 설정
param_grid = {
    'n_neighbors': [6, 7],  # 이웃의 수
    'weights': ['uniform', 'distance'],  # 가중치 방식
    'metric': ['euclidean', 'manhattan']  # 거리 계산 방식
}
```

```python
# GridSearchCV 설정: 모델, 파라미터 그리드, 교차검증 횟수(cv)
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='accuracy')
```

```python
from sklearn.ensemble import RandomForestClassifier

clf_rf = RandomForestClassifier(max_depth=2, random_state=0)
clf_rf.fit(X_train, y_train)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

clf_svc = make_pipeline(StandardScaler(), SVC(gamma='auto'))
clf_svc.fit(X_train, y_train)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.datasets import make_classification

clf_extra = ExtraTreesClassifier(n_estimators=100, random_state=42)
clf_extra.fit(X_train, y_train)
```

```python
from sklearn.ensemble import VotingClassifier

eclf1 = VotingClassifier(estimators=[('svc', clf_svc), ('rf', clf_rf), ('extra', clf_extra)])
eclf1.fit(X_train, y_train)
```

```python
from sklearn.ensemble import VotingClassifier

eclf2 = VotingClassifier(voting = 'soft', estimators=[('svc', clf_svc), ('rf', clf_rf), ('extra', clf_extra)])
eclf2.fit(X_train, y_train)
```
