```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', as_frame=False)
```

```python
X, y = mnist.data, mnist.target
```

```python
X_train, X_test, y_train, y_test = X[:50000], X[50000:60000],y[:50000], y[50000:60000],
```

```python
from sklearn.ensemble import RandomForestClassifier

rnd_clf = RandomForestClassifier(n_estimators=500, max_leaf_nodes=16, n_jobs=-1, random_state=42)
rnd_clf.fit(X_train, y_train)
```

```python
rnd_clf.score(X_test, y_test)
```

```python
from sklearn.tree import ExtraTreeClassifier

ext_clf = ExtraTreeClassifier(random_state=0)
ext_clf.fit(X_train, y_train)
```

```python
ext_clf.score(X_test, y_test)
```

```python
from sklearn.svm import SVC

svm_clf = SVC(random_state=42)
svm_clf.fit(X_train, y_train)
```

```python
svm_clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import VotingClassifier

voting_clf = VotingClassifier(
    estimators = [
        ('rf', RandomForestClassifier(n_estimators=500, max_leaf_nodes=16, n_jobs=-1, random_state=42)),
        ('et', ExtraTreeClassifier(random_state=0)),
        ('svc', SVC(random_state=42))
    ]
)
voting_clf.fit(X_train, y_train)
```

```python
voting_clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import StackingClassifier

stacking_clf = StackingClassifier(
    estimators = [
        ('rf', RandomForestClassifier(n_estimators=500, max_leaf_nodes=16, n_jobs=-1, random_state=42)),
        ('et', ExtraTreeClassifier(random_state=0)),
        ('svc', SVC(random_state=42))
    ],
    final_estimator=RandomForestClassifier(random_state=42),
    cv=5
)
stacking_clf.fit(X_train, y_train)
```

```python
stacking_clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier

ada_clf = AdaBoostClassifier(
    SVC(random_state=42), n_estimators=30, learning_rate=0.5, random_state=42)
ada_clf.fit(X_train, y_train)
```

```python
ada_clf.score(X_test, y_test)
```
