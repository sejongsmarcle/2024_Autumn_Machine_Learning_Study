```python
from sklearn.datasets import fetch_openml

# 사이킷런 1.2에서 추가된 parser 매개변수 기본값이 1.4 버전에서 'liac-arff'에서 'auto'로 바뀝니다.
# 'auto'일 경우 희소한 ARFF 포맷일 때는 'liac-arff', 그렇지 않을 때는 'pandas'가 됩니다.
# 이에 대한 경고를 피하려면 parser='auto'로 지정하세요.
mnist = fetch_openml('mnist_784', as_frame=False)
```

```python
X, y = mnist['data'], mnist['target']
```

```python
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
sgd_clf = SGDClassifier(random_state=42)
%time sgd_clf.fit(X_train, y_train)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=154, svd_solver="randomized",random_state=42)
X_reduced_train = pca.fit_transform(X_train)
X_reduced_test = pca.transform(X_test)
```

```python
sgd_clf = SGDClassifier(random_state=42, max_iter=50)
%time sgd_clf.fit(X_reduced_train, y_train)
```

```python
score = sgd_clf.score(X_reduced_test, y_test)
print(f"Model score: {score}")
```

```python
sgd_clf = SGDClassifier(random_state=42, max_iter=100)
%time sgd_clf.fit(X_reduced_train, y_train)
```

```python
score = sgd_clf.score(X_reduced_test, y_test)
print(f"Model score: {score}")
```

```python
sgd_clf = SGDClassifier(random_state=42, max_iter=500)
%time sgd_clf.fit(X_reduced_train, y_train)
```

```python
score = sgd_clf.score(X_reduced_test, y_test)
print(f"Model score: {score}")
```
